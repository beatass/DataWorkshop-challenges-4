```
def preprocessing(sentence):
    return [word for word in simple_preprocess(sentence) \
           if word not in STOPWORDS]

def read_sentences(filename):
    with open(filename, 'rb') as f:
        for line in f:
            yield preprocessing(line)
```

```
%time sentences = list(read_sentences('reviews_data.txt'))
```

#Topic modeling - LDA

```
sentences_light = np.random.permutation(sentences) #now we want to mix the words so that there is no relationship between them
```

```
sentences_light = sentences_light[:1000] #tell us how many sentences we want to finally take
```

```
%time dictionary = Dictionary(sentences_light) #now we are building a dictionary
```

```
len(dictionary)
```

```
%time bow_corpus = [dictionary.doc2bow(sent) for sent in sentences_light]
```

```
dictionary.doc2bow(['car'])
```

```
dictionary.token2id['car']
```

```
%time lda_model = LdaMulticore(bow_corpus, id2word=dictionary, num_topics=100, passes=20, workers=8)
```

```
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

#Visualisation of topics

```
lda_vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
pyLDAvis.display(lda_vis)
```
